```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Data_Cortex_Nuclear.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
df_train = df.sample(frac=0.95, random_state=123)
```

```python
df_train.shape
```

```python
df_test = df.drop(df_train.index)
```

```python
df_test.shape
```

```python
df_train.reset_index(drop=True, inplace=True)
df_test.reset_index(drop=True, inplace=True)
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
from pycaret.anomaly import *
```

```python
anomaly = setup(data = df_train, normalize=True,
                ignore_features=["MouseID"], session_id=123)
```

```python
iforest = create_model('iforest')
```

```python
histogram = create_model('histogram')
```

```python
print(histogram)
```

```python
plot_model(histogram)
```

```python
evaluate_model(histogram)
```

```python
print(iforest)
```

```python
plot_model(iforest)
```

```python
evaluate_model(iforest)
```

```python
models()
```

```python
iforest_results = assign_model(iforest)
histogram_results = assign_model(histogram)
```
